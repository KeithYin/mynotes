# chapter-02 Probability
> Probability theory is nothing but common sense reduced to calculation

* probability 的两种解释方式
    * frequentist interpretation: probabilities represent long run frequencies of events (即: 实验很多次,事件发生的概率)
    * Bayesian interpretation: is used to quantify our **uncertainty** about something. 表示的是信息(不确定性), 而不是多次实验
    * 频率派的解释是  多次实验的结果(实验的解释, 如果不实验, 此解释无用)
    * 贝叶斯派的解释是 下次实验,出现哪个事件的概率

### 概率的几个重要性质
**union of two events**
$$p(A or B) = p(A) + p(B) - p(A and B)$$

**joint probabilities**
p(A, B) = p(A and B) = p(B|A)p(A)

**continuous variable**
* 概率是定义在区间上的. 如果不是区间, 那就表示的是概率密度
* $F(q) = p(X \le q)$, $F(q)$ 的含义是, 随机连续变量 $X\le q$的概率是多少
* $f(x) = \frac{d}{dx}F(x)$ 表示是概率密度函数, 其值是可能**大于1**的!!!
* $P(x\le X \le x+dx) \approx p(x)dx$, 其中 $p(x)$ 为概率密度函数, $P(...)$ 为概率

**分位数 quantiles**
* $P(X\le x_\alpha)=\alpha, F^{-1}(\alpha) = x_\alpha$, $x_\alpha$ 称为 $\alpha ~~~quantiles$, 其中 $\alpha$ 表示累积概率,   $x_\alpha$ 表示 累积到哪个位置.

### 均值方差计算

大写 $X$ 为随机变量, 均值方差的公式是对于随机变量说的. 小写 $x$ 表示真实的采样值 

* 均值
$$
\Bbb E(X) = \sum_{x\in \mathcal X} x*p(x)
$$

* 方差
$$
Var(X) = \Bbb E[(X-\mu)^2] = \int (x-\mu)^2p(x)dx = \Bbb E[X^2] - \mu^2
$$

### 几个重要分布
**离散变量**

* 二项分布 (binomial distribution): N 次 Bernoulli 实验(扔硬币实验), 正例出现 K 次的概率是多少. 已知 一次 Bernoulli 实验, 正例出现的概率$\theta$
    * k 次正面朝上的概率为 (k 次正面, n-k 次反面), 有 $C_n^k$ 种组合方式, 所以概率可以写成下面形式

$$
Bin(k|n,\theta) = C_n^k \theta^k(1-\theta)^{n-k}
$$

* multinomial distribution (多项式分布): N 次 扔 K-sided dice 实验, 第k面出现的概率为 $\theta_k$
    * 令 $\mathbf x=\{x_1, x_2, ..., x_K\}$ 为随机变量, 其中 $x_k$ 为 N 次实验中, 第$k$面出现的次数
    * $x_1$ 次 1 朝上, $x_2$次2朝上, ... 概率为 $\prod_{k=1}^K\theta_k^{x_k}$, 有 $C_n^{x_1}C_{n-x_1}^{x_2}C_{n-x_1-x_2}^{x_3}... = \frac{n!}{x_1!x_2!...x_K!}$ 种组合方式
    * 一次实验得到的分布为 categorical/discrete/multinoulli distribution. 三个名字,哪个都行. multinoulli对英语 bernoulli
$$
Mu(\mathbf x|n, \theta) = \frac{n!}{x_1!x_2!...x_K!} \prod_{k=1}^K\theta_k^{x_k}
$$

* Poisson Distribution
    * 随机变量 $X \in \{0, 1, 2, ...\}$ 符合 poisson distribution, 那就满足下面公式. The Poisson distribution is often used as a model for counts of rare events like radioactive decay and traffic accidents
$$
Poi(x|\lambda) = e^{-\lambda}\frac{\lambda^x}{x!}
$$
